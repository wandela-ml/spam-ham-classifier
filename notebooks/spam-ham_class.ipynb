```python
import pandas as pd
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('spam.csv', encoding='latin1')
df = df[['v1', 'v2']]
df.columns = ['Target', 'Email']
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['Email'], df['Target'], test_size=0.2, random_state=42, stratify = df['Target'])
```

```python
vector = TfidfVectorizer(stop_words='english')
X_train_tfidf = vector.fit_transform(X_train)
X_test_tfidf = vector.transform(X_test)
```

# Train the model

```python
model = LogisticRegression()
```

```python
model.fit(X_train_tfidf, y_train)
```

```python
predictions = model.predict(X_test_tfidf)
```

```python
print("Accuracy Score: ", accuracy_score(y_test, predictions))
print(classification_report(y_test, predictions))
```

```python
mesage = ['Congrations! Claim your free prize today']
mesage_vect = vector.transform(mesage)
prediction = model.predict(mesage_vect)
```

```python
print(prediction)
```

# Find important spam words

```python
feature_names = vector.get_feature_names_out()
coefficients = model.coef_[0]
```

# Create a dataframe

```python
importance = pd.DataFrame({
    'Word': feature_names,
    'Weight': coefficients
})
```

# Most spam indicating words

```python
importance.sort_values(
    by='Weight',
    ascending = False 
).head(20)
```
